```python
import pandas as pd #Load Pandas
olympicMedals=pd.read_csv('/content/drive/MyDrive/DATA/olympic_medals.csv')

#2.A print the size of the dataframe (number of columns and rows)
print('Number of Rows ',olympicMedals.shape[0])
print('Number of Columns ', olympicMedals.shape[1],end='\n\n')
#2.B. print the datatype for all columns
print('Data Columns')
print(olympicMedals.info(),end='\n\n')
#2.C. print unique cities that have won medals
print('Unique Cities', olympicMedals['Location'].unique(),end='\n\n')
#2.D. How many medals has Team USA won
print('Number of Medals won by USA',len(olympicMedals[olympicMedals['Nationality']=='USA']),end='\n\n')
#2.E. print total number of medla per nationality
print("Medals won by Nationality")
with pd.option_context('display.max_rows', None):
    print(olympicMedals.groupby(by='Nationality')['Medal'].count().to_markdown)
```
